```python
import matplotlib.pyplot as plt
import numpy as np

from pydicom import dcmread
import pydicom
```

```python
def show_slice_window(slice, level, window):
    
    """
    Function to display an image slice
    Input is a numpy 2D array
    """
    
    max = level + window/2
    min = level - window/2
    
    slice = slice.clip(min,max)
    plt.figure()
    plt.imshow(slice, cmap="gray")
    #plt.savefig('L'+str(level)+'W'+str(window))
```

```python
ds = dcmread("dicom_dir/ID_0000_AGE_0060_CONTRAST_1_CT.dcm")
```

```python
img = ds.pixel_array
```

```python
slope = ds.RescaleSlope
```

```python
intercepte = ds.RescaleIntercept
```

```python
hu_image = img * slope + intercepte
```

```python
show_slice_window(hu_image, 120, 30)
```

```python
show_slice_window(hu_image, 80, 40)
```
